```python
# --- Import libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load dataset ---
df = pd.read_csv("converted_file (1).csv")

# --- Basic info ---
print("Shape of dataset:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum().sum())

# --- Preview dataset ---
display(df.head())

# --- Statistical summary ---
display(df.describe(include='all'))

# --- Check for duplicates ---
print("Number of duplicate rows:", df.duplicated().sum())

# --- Correlation heatmap (for numeric columns) ---
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# --- Distribution plots for numeric columns ---
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols].hist(figsize=(15, 12), bins=20)
plt.suptitle("Distributions of Numeric Columns")
plt.show()

# --- Check categorical columns ---
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols[:5]:  # show first 5 for brevity
    plt.figure(figsize=(8, 4))
    sns.countplot(y=col, data=df)
    plt.title(f"Count Plot - {col}")
    plt.show()

# --- Preprocessing Example ---
# Encode categorical variables
df_encoded = pd.get_dummies(df, drop_first=True)

# Normalize numerical columns (if needed)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

print("\nPreprocessing complete! Final dataset shape:", df_encoded.shape)
display(df_encoded.head())
```

```python
# --- Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load your dataset ---
df = pd.read_csv("converted_file (1).csv")

# --- Compute correlation matrix ---
corr_matrix = df.corr(numeric_only=True)

# --- Plot correlations for each grade variable ---
targets = ['G1', 'G2', 'G3']

for target in targets:
    if target in corr_matrix.columns:
        plt.figure(figsize=(8, 5))
        corr_values = corr_matrix[target].sort_values(ascending=False).drop(target)
        sns.barplot(x=corr_values.values, y=corr_values.index, palette="viridis")
        plt.title(f"Correlation of Attributes with {target}", fontsize=14)
        plt.xlabel("Correlation Coefficient", fontsize=12)
        plt.ylabel("Attributes", fontsize=12)
        plt.grid(alpha=0.3)
        plt.show()
```

```python
# =====================================
# ðŸ”¹ Find and Visualize Strongly Correlated Features
# =====================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Compute correlation matrix (numerical features only)
corr_matrix = df.corr(numeric_only=True)

# Display heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of All Numerical Features")
plt.show()

# Identify pairs with strong correlation (|r| > 0.7)
threshold = 0.7
strong_corr = []

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            strong_corr.append((corr_matrix.columns[i],
                                corr_matrix.columns[j],
                                corr_matrix.iloc[i, j]))

# Convert to DataFrame for clarity
strong_corr_df = pd.DataFrame(strong_corr, columns=['Feature 1', 'Feature 2', 'Correlation'])
strong_corr_df = strong_corr_df.sort_values(by='Correlation', ascending=False)

print("\nðŸ” Strongly Correlated Feature Pairs (|r| > 0.7):\n")
display(strong_corr_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# --- 1. Define File Name ---
# This code assumes 'converted_file (1).csv' is already uploaded
# to your Colab session (e.g., via the 'Files' sidebar).
file_name = 'converted_file (1).csv'

# --- 2. Main Plotting Code ---
try:
    # Load the dataset directly
    df = pd.read_csv(file_name)
    print(f"Successfully loaded {file_name}")

    # Set the visual style for the plots
    sns.set_theme(style="whitegrid")

    # Create a 3x2 grid of subplots for all 6 visualizations
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 18))

    # Add a main title for the entire figure
    fig.suptitle('Visualizing Six Key Correlations in Student Data (Revised)', fontsize=22, y=1.03)

    # --- Plot 1: Medu vs Fedu (Box Plot) ---
    sns.boxplot(
        data=df,
        x='Medu',
        y='Fedu',
        ax=axes[0, 0]
    )
    axes[0, 0].set_title("Father's Education by Mother's (Corr: +0.63)", fontsize=14)
    axes[0, 0].set_xlabel("Mother's Education (Medu)")
    axes[0, 0].set_ylabel("Father's Education (Fedu)")

    # --- Plot 2: famrel vs goout (Box Plot) ---
    sns.boxplot(
        data=df,
        x='famrel',
        y='goout',
        ax=axes[0, 1]
    )
    axes[0, 1].set_title('Going Out by Family Relations (Corr: -0.42)', fontsize=14)
    axes[0, 1].set_xlabel('Quality of Family Relations (famrel)')
    axes[0, 1].set_ylabel('Going Out with Friends (goout)')

    # --- Plot 3: age vs failures (Box Plot) ---
    sns.boxplot(
        data=df,
        x='failures',
        y='age',
        ax=axes[1, 0]
    )
    axes[1, 0].set_title('Age Distribution by Past Failures (Corr: +0.33)', fontsize=14)
    axes[1, 0].set_xlabel('Number of Past Failures (failures)')
    axes[1, 0].set_ylabel('Age')

    # --- Plot 4: traveltime vs absences (Box Plot - Outliers Hidden) ---
    sns.boxplot(
        data=df,
        x='traveltime',
        y='absences',
        ax=axes[1, 1],
        showfliers=False  # <-- MODIFICATION: Hide outliers for better scaling
    )
    axes[1, 1].set_title('Absences by Travel Time (Corr: +0.29, Outliers Hidden)', fontsize=14)
    axes[1, 1].set_xlabel('Home-to-School Travel Time (traveltime)')
    axes[1, 1].set_ylabel('Number of Absences (absences)')

    # --- Plot 5: studytime vs failures (Heatmap) ---
    # Create a contingency table (crosstab)
    study_fail_crosstab = pd.crosstab(df['studytime'], df['failures'])

    sns.heatmap(
        study_fail_crosstab,
        annot=True,      # Show the counts in each cell
        fmt='d',         # Format as an integer
        cmap='YlGnBu',   # Use a color map
        linewidths=0.5,
        ax=axes[2, 0]
    )
    axes[2, 0].set_title('Count by Study Time & Failures (Corr: -0.24)', fontsize=14)
    axes[2, 0].set_xlabel('Number of Past Failures (failures)')
    axes[2, 0].set_ylabel('Weekly Study Time (studytime)')

    # --- Plot 6: health vs absences (Box Plot - Outliers Hidden) ---
    sns.boxplot(
        data=df,
        x='health',
        y='absences',
        ax=axes[2, 1],
        showfliers=False  # <-- MODIFICATION: Hide outliers for better scaling
    )
    axes[2, 1].set_title('Absences by Health Status (Corr: +0.19, Outliers Hidden)', fontsize=14)
    axes[2, 1].set_xlabel('Current Health Status (health)')
    axes[2, 1].set_ylabel('Number of Absences (absences)')

    # --- Final Touches ---
    # Adjust layout to prevent titles/labels from overlapping
    plt.tight_layout(rect=[0, 0.03, 1, 0.98])

    # Save the figure (optional, but good practice)
    plt.savefig('all_correlations_grid_scaled.png', dpi=300)

    # Display the plot
    plt.show()

except FileNotFoundError:
    print(f"\n--- ERROR ---")
    print(f"File '{file_name}' not found.")
    print("Please make sure you have uploaded it to the Colab 'Files' sidebar.")
except KeyError as e:
    print(f"\n--- PLOTTING ERROR ---")
    print(f"A column was not found in the file: {e}")
    print("Please check that your CSV file has the correct column names.")
```

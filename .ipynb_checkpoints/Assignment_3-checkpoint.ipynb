```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import os
```

```python
DATA_PATH = "titanic_train_dc904cdc-94af-4bfb-b2b8-240e0af633f7.csv"

df = pd.read_csv(DATA_PATH)
print("Shape:", df.shape)
print(df.head(3))
print("\nNull counts:\n", df.isna().sum())
```

Running the dataset and looking for missing values

```python
print("Dataset info:\n")
print(df.info())

print("\nNumeric summary:\n")
print(df.select_dtypes(include=["int64", "float64"]).describe().T)

print("\nCategorical summary:\n")
print(df.select_dtypes(include=["object"]).describe().T)

print("\nMissing values per column:\n")
print(df.isna().sum())

print("\nRanges for numeric columns:\n")
print(df.select_dtypes(include=["int64", "float64"]).max() - df.select_dtypes(include=["int64", "float64"]).min())
```

Exploring the dataset

```python
os.makedirs("figs", exist_ok=True)

# Histograms for numeric attributes
num_cols = [c for c in df.columns if df[c].dtype in ["int64", "float64"] and c not in ["PassengerId"]]
for col in num_cols:
    plt.figure()
    df[col].dropna().hist(bins=30)
    plt.title(f"Histogram: {col}")
    plt.xlabel(col); plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(f"figs/hist_{col}.png", dpi=150)
    plt.show()

# Boxplots of Age & Fare by Survived
for col in ["Age", "Fare"]:
    if col in df.columns and "Survived" in df.columns:
        plt.figure()
        groups = [
            df.loc[df["Survived"] == 0, col].dropna(),
            df.loc[df["Survived"] == 1, col].dropna()
        ]
        plt.boxplot(groups, labels=["Not Survived (0)", "Survived (1)"], showmeans=True)
        plt.title(f"Boxplot: {col} by Survived")
        plt.ylabel(col)
        plt.tight_layout()
        plt.savefig(f"figs/box_{col}_by_survived.png", dpi=150)
        plt.show()

# Scatter: Age vs Fare, colored by Survived
if all(c in df.columns for c in ["Age", "Fare", "Survived"]):
    plt.figure()
    mask0 = df["Survived"] == 0
    mask1 = df["Survived"] == 1
    plt.scatter(df.loc[mask0, "Age"], df.loc[mask0, "Fare"], alpha=0.6, label="Not Survived (0)")
    plt.scatter(df.loc[mask1, "Age"], df.loc[mask1, "Fare"], alpha=0.6, label="Survived (1)")
    plt.xlabel("Age"); plt.ylabel("Fare"); plt.title("Scatter: Age vs Fare by Survival")
    plt.legend()
    plt.tight_layout()
    plt.savefig("figs/scatter_age_fare_by_survival.png", dpi=150)
    plt.show()

# Q-Q plots for Age & Fare
for col in ["Age", "Fare"]:
    if col in df.columns:
        data = df[col].dropna()
        if len(data) > 3:
            plt.figure()
            stats.probplot(data, dist="norm", plot=plt)
            plt.title(f"Q-Q Plot: {col}")
            plt.tight_layout()
            plt.savefig(f"figs/qq_{col}.png", dpi=150)
            plt.show()

# Bar charts: survival rate by Sex, Pclass, Embarked
def plot_survival_rate(col, title, fname):
    rates = df.groupby(col)["Survived"].mean().sort_values(ascending=False)
    plt.figure()
    rates.plot(kind="bar", color="skyblue", edgecolor="black")
    plt.ylabel("Survival Rate")
    plt.title(title)
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.savefig(f"figs/{fname}.png", dpi=150)
    plt.show()

if "Sex" in df.columns:
    plot_survival_rate("Sex", "Survival Rate by Sex", "bar_survival_by_sex")

if "Pclass" in df.columns:
    plot_survival_rate("Pclass", "Survival Rate by Pclass", "bar_survival_by_pclass")

if "Embarked" in df.columns:
    tmp = df.copy()
    tmp["Embarked"] = tmp["Embarked"].fillna(tmp["Embarked"].mode().iloc[0])
    plot_survival_rate("Embarked", "Survival Rate by Embarked", "bar_survival_by_embarked")
```

```python
def summarize_survival_rate(group_col):
    """Prints survival counts and rates by a given column."""
    summary = df.groupby(group_col)["Survived"].agg(
        count="count",
        survived_sum="sum",
        survival_rate="mean"
    ).sort_values("survival_rate", ascending=False)
    print(f"\nSurvival by {group_col}:")
    print(summary)
    return summary

# Overall survival rate
overall_rate = df["Survived"].mean()
print(f"Overall survival rate: {overall_rate:.2%}")

# Survival by Sex
if "Sex" in df.columns:
    sex_summary = summarize_survival_rate("Sex")

# Survival by Pclass
if "Pclass" in df.columns:
    pclass_summary = summarize_survival_rate("Pclass")

# Survival by Age groups
if "Age" in df.columns:
    bins = [0, 10, 18, 30, 45, 60, 120]
    labels = ["0-9", "10-17", "18-29", "30-44", "45-59", "60+"]
    df["AgeBand"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)
    age_summary = summarize_survival_rate("AgeBand")

# Survival by Embarked
if "Embarked" in df.columns:
    df["Embarked_filled"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
    embarked_summary = summarize_survival_rate("Embarked_filled")

# Combined analysis: Sex x Pclass
if set(["Sex", "Pclass"]).issubset(df.columns):
    print("\nSurvival rate by Sex x Pclass:")
    pivot_sex_pclass = df.pivot_table(values="Survived", index="Sex", columns="Pclass", aggfunc="mean")
    print(pivot_sex_pclass)

# Combined analysis: Sex x AgeBand
if set(["Sex", "AgeBand"]).issubset(df.columns):
    print("\nSurvival rate by Sex x AgeBand:")
    pivot_sex_age = df.pivot_table(values="Survived", index="Sex", columns="AgeBand", aggfunc="mean")
    print(pivot_sex_age)
```

```python

prep = df.copy()


# filled missing values with mode
if "Embarked" in prep.columns:
    embarked_mode = prep["Embarked"].mode().iloc[0]
    prep["Embarked"] = prep["Embarked"].fillna(embarked_mode)

# Cabin: ~77% missing  keep an indicator column & drop raw Cabin
if "Cabin" in prep.columns:
    prep["CabinKnown"] = prep["Cabin"].notna().astype(int)
    prep = prep.drop(columns=["Cabin"])

# Age: impute with Sex  Pclass median (class-wise imputation)
if set(["Age","Sex","Pclass"]).issubset(prep.columns):
    age_group_median = prep.groupby(["Sex","Pclass"])["Age"].transform("median")
    prep["Age_imputed"] = prep["Age"].fillna(age_group_median)
    # fallback to global median if still missing
    prep["Age_imputed"] = prep["Age_imputed"].fillna(prep["Age"].median())
else:
    prep["Age_imputed"] = prep["Age"].fillna(prep["Age"].median())

# Fare: fill missing (rare) with Pclass median
if "Fare" in prep.columns:
    if prep["Fare"].isna().any():
        if "Pclass" in prep.columns:
            fare_group_median = prep.groupby("Pclass")["Fare"].transform("median")
            prep["Fare"] = prep["Fare"].fillna(fare_group_median)
        prep["Fare"] = prep["Fare"].fillna(prep["Fare"].median())

# noise/outliers
def iqr_cap(series, k=1.5):
    s = series.dropna()
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lo, hi = q1 - k*iqr, q3 + k*iqr
    return series.clip(lower=lo, upper=hi)

if "Age_imputed" in prep.columns:
    prep["Age_cap"] = iqr_cap(prep["Age_imputed"])
if "Fare" in prep.columns:
    prep["Fare_cap"] = iqr_cap(prep["Fare"])

# transformation
if "Fare" in prep.columns:
    prep["Fare_log1p"] = np.log1p(prep["Fare"])

# feature engineering-

# Extract Title from Name
if "Name" in prep.columns:
    prep["Title"] = prep["Name"].str.extract(r",\s*([^\.]+)\.", expand=False).str.strip()
    # group rare titles
    title_counts = prep["Title"].value_counts()
    rare_titles = title_counts[title_counts < 10].index
    prep["Title"] = prep["Title"].replace(rare_titles, "Rare")

# Family features
if set(["SibSp","Parch"]).issubset(prep.columns):
    prep["FamilySize"] = prep["SibSp"] + prep["Parch"] + 1
    prep["IsAlone"] = (prep["FamilySize"] == 1).astype(int)

# Encoding categorical variables

# Sex  binary
if "Sex" in prep.columns:
    prep["Sex_bin"] = prep["Sex"].map({"male": 0, "female": 1}).astype(int)

# One-hot encode Pclass, Embarked, Title
categoricals = []
if "Pclass" in prep.columns: categoricals.append("Pclass")
if "Embarked" in prep.columns: categoricals.append("Embarked")
if "Title" in prep.columns: categoricals.append("Title")

prep = pd.get_dummies(prep, columns=categoricals, drop_first=True)

# Scaling (z-score) for numeric features

scale_cols = [c for c in ["Age_cap","Fare_cap","Fare_log1p","SibSp","Parch","FamilySize"] if c in prep.columns]

try:
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    if scale_cols:
        prep[[f"{c}_z" for c in scale_cols]] = scaler.fit_transform(prep[scale_cols])
except ImportError:
    print(" sklearn not available; skipping scaling")

# Final cleaned dataset

drop_cols = [c for c in ["PassengerId","Name","Ticket"] if c in prep.columns]
model_df = prep.drop(columns=drop_cols, errors="ignore").copy()

print("Preprocessing complete. Preview:")
print(model_df.head())

# Save cleaned dataset
clean_path = "titanic_cleaned.csv"
model_df.to_csv(clean_path, index=False)
print(f" Cleaned data saved to: {clean_path}")
```

```python

print("========== KEY INSIGHTS ==========\n")

# Overall survival rate
overall_rate = df["Survived"].mean()
print(f"Overall survival rate: {overall_rate:.2%}")

# Sex
if "Sex" in df.columns:
    sex_rate = df.groupby("Sex")["Survived"].mean()
    print("\nSurvival rate by Sex:")
    print(sex_rate.to_string())

# Pclass
if "Pclass" in df.columns:
    pclass_rate = df.groupby("Pclass")["Survived"].mean().sort_values(ascending=False)
    print("\nSurvival rate by Pclass:")
    print(pclass_rate.to_string())

# Age bands
if "Age" in df.columns:
    bins = [0, 10, 18, 30, 45, 60, 120]
    labels = ["0-9", "10-17", "18-29", "30-44", "45-59", "60+"]
    df["AgeBand"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)
    age_rate = df.groupby("AgeBand")["Survived"].mean().sort_values(ascending=False)
    print("\nSurvival rate by Age Band:")
    print(age_rate.to_string())

# Embarked
if "Embarked" in df.columns:
    emb_mode = df["Embarked"].mode().iloc[0]
    embarked_rate = df["Embarked"].fillna(emb_mode).groupby(df["Embarked"]).apply(lambda x: df.loc[x.index, "Survived"].mean())
    print("\nSurvival rate by Embarked:")
    print(embarked_rate.to_string())

# Combined (Sex x Pclass)
if set(["Sex", "Pclass"]).issubset(df.columns):
    print("\nSurvival rate by Sex  Pclass:")
    print(df.pivot_table(values="Survived", index="Sex", columns="Pclass", aggfunc="mean"))

# Combined (Sex x AgeBand)
if set(["Sex", "AgeBand"]).issubset(df.columns):
    print("\nSurvival rate by Sex  AgeBand:")
    print(df.pivot_table(values="Survived", index="Sex", columns="AgeBand", aggfunc="mean"))

print("\n========== PREPROCESSING REPORT ==========\n")

# Missing values before preprocessing
missing_before = df.isna().sum()

# Missing values after preprocessing
missing_after = model_df.isna().sum()

print("Missing values (before):")
print(missing_before[missing_before > 0])

print("\nMissing values (after):")
print(missing_after[missing_after > 0] if missing_after.sum() > 0 else "No missing values left ")

print("\nTransformations performed:")
print("- Age: ~20% missing  imputed using Sex  Pclass median")
print("- Cabin: ~77% missing  replaced with binary indicator (CabinKnown)")
print("- Embarked: 2 missing  filled with mode (most common value)")
print("- Fare: missing values (if any)  filled with Pclass median")
print("- Outliers: Age and Fare capped using IQR method")
print("- Fare also transformed using log(1+Fare) to reduce skewness")
print("- Feature Engineering: extracted Title from Name, created FamilySize & IsAlone")
print("- Encoding: Sex  binary, Pclass/Embarked/Title  one-hot encoding")
print("- Scaling: applied z-score standardization to Age, Fare, SibSp, Parch, FamilySize")

print("\n Preprocessing complete. Cleaned dataset saved as titanic_cleaned.csv")
```
